### Part B (0.6)

```python
lis1 = ['DS', 123, 'Stats', 'prob', 56.9, True, '.', 'DS']
lis2 = ['pi', 'alpha', 452, 9582, 123, 'prob', 'hello world', '.', 'DS']
```

###### Sol1 (70% of the complete score)

```python
%%timeit
lis1_new = []
lis2_new = []

for word in lis1:
    if word not in lis1_new:
        lis1_new.append(word)
for word in lis2:
    if word not in lis2_new:
        lis2_new.append(word)

common=[]
for word in lis1_new:
    if word in lis2_new:
        common.append(word)
```

```python
print(len(common) , common)
```

###### Sol2 (complete score)

```python
%%timeit

common_rep=[]
common=[]
for word in lis1:
    if word in lis2:
        common_rep.append(word)

# making the values in common list unique
for word in common_rep:
    if word not in common:
        common.append(word)
```

```python
print(len(common) , common)
```

###### Sol3 (complete score)

```python
%%timeit

common=[]
for word in lis1:
    if (word in lis2) & (word not in common):
        common.append(word)
```

```python
print(len(common), common) 
```

# Question3 (1)

###### Sol1 (80% of the complete score)

```python
%%timeit
lis = [ 123, 36, 43, 76, 68, 19, 41, 82, 72, 92, 9, 26]
prime_list = []

for num in lis:
    if (num > 1) and (int(num) == num):
        num = int(num)
        flag = True
        all_nums = range(2, int(num**0.5 + 1))
        for i in all_nums:
            if num % i == 0:
                flag = False
        if flag:
            prime_list.append(num)
```

```python
print(prime_list)
```

###### Sol2 (complete score)

```python
%%timeit
lis = [1, 2.0, 2.01, 36, 43, 76, 68, 19, 41, 82, 72, 92, 9, 26]
prime_list = []

for num in lis:
    if (num > 1) and (int(num) == num):
        num = int(num)
        flag = True
        max_num = int(num**0.5 + 1)
        i=2
        
        while flag and (i<max_num):
            if num % i ==0:
                flag=False
            i+=1
        if flag:
            prime_list.append(num)
```

### Part B (0.7)

```python
lis = ['acdooaaesdfe', 'acbbac', 'sad', 'ooaaesdfe', 'soidfle', 'sskjds', 'sfsdf', 'ksrj']
```

###### Sol1 (80% of the complete score)

```python
%%timeit
flag = False
for word in lis:
    counter=0
    while (not flag) and counter<(len(lis)):
        if (word in lis[counter]) and (word != lis[counter]):
            flag = True
        counter+=1
```

```python
if flag:
    print(True)
else:
    print(False)
```

###### Sol2 (80% of the complete score)

```python
%%timeit
flag=True
i=0
while flag and i<len(lis)-1:
    for j in range(i+1,len(lis)):
        if (lis[i] in lis[j] or lis[j] in lis[i]) & (lis[i] != lis[j]):
            flag=False
    i += 1
```

```python
if flag:
    print(False)
else:
    print(True)
```

###### Sol3 (complete score)

```python
%%timeit
flag = False
for w1 in lis:
    for w2 in lis:
        if (w1 in w2) and (w1 != w2):
            flag = True
```

```python
if flag:
    print(True)
else:
    print(False)
```
